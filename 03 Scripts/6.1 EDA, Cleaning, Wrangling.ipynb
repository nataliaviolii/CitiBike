‘gender' is written numerically and needs to be categorized. In this case I will derive a new column:
- 1 as 'male' 
- 2 as ‘female’
- 0 as ‘unknown

```python
def gender_cat(gender):
    if gender == 1:
        return 'Male'
    elif gender == 2:
        return 'Female'
    else:
        return 'Unknown'
```

```python
df['gender_category'] = df['gender'].apply(gender_cat)
```

```python
# The trip duration is written in seconds and I decided to create a function to convert the value to minutes
def time_convert(seconds):
  mins = seconds//60
  return mins  
```

```python
df['trip_duration_min'] = df['trip_duration'].apply(time_convert)
```

```python
df.head()
```

```python
df['trip_duration_min'].describe()
```

```python
def time_convertt(secs):
  days = secs//86400
  hours = (secs - days*86400)//3600
  mins = (secs - days*86400 - hours*3600)//60
  return mins 
```

```python
df['trip_duration_min1'] = df['trip_duration'].apply(time_convertt)
```

```python
df.head()
```

```python
df['trip_duration_min1'].describe()
```
