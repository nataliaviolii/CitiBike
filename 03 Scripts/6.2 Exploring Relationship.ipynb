## Exploring Relationship

### Correlation (Heatmap)

```python
df.dtypes
```

```python
# Create a subset excluding some unnecessary data
subset = df[['start_hour', 'start_station_id', 'end_station_id', 'birth_year', 'gender', 'trip_duration_min']]
```

```python
subset.head()
```

```python
# Create a correlation matrix using pandas
subset.corr()
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(8,8))


# Create a correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above
corr = sns.heatmap(subset.corr(), annot = True, ax = ax)
```

- 0: no relationship
- 0.1–0.3: weak relationship
- 0.3–0.5: moderate relationship
- 0.5–1.0: strong relationship

#### Observation:
1. trip_duration_min with start_hour = 0.041 shows a very weak positive correlation (or almost no correlation) --> an earlier start time does not really affect the length of the trip duration
2. trip_duration_min with start_station_id = 0.062 shows a very weak positive correlation (or almost no correlation) --> the station where the user can start cycling does not really affect the length of the trip duration
3. trip_duration_min with end_station_id = 0.059 shows a very weak positive correlation (or almost no correlation) --> the station where the user finishes cycling does not really affect the length of the trip duration
4. trip_duration_min with birth_year = -0.047 shows a very weak negative correlation (or almost no correlation) --> the older the user does not mean the shorter the trip duration
5. trip_duration_min with gender = -0.16 shows a weak negative correlation (compared to other variables, this variable has a higher value even though the relationship itself is still weak). 0 = unknown, 1 = male, 2 = female --> unknown users appear to have longer trip lengths than other types of users

```python
# Checking the result no. 4 (trip_duration_min and gender)
df.groupby('gender').agg({'trip_duration_min' : ['mean', 'min', 'max']})
```

Unknown users appear to have longer trip lengths than other types of users

```python
# Processing the start & end date time
# Adding more columns for analysis purposes

# start time
df.start_time = pd.to_datetime(df.start_time, format='%Y-%m-%d %H:%M:%S')
df['start_year'] = df.start_time.apply(lambda x: x.year)
df['start_month'] = df.start_time.apply(lambda x: x.month)
df['start_week'] = df.start_time.apply(lambda x: x.week)
df['start_day'] = df.start_time.apply(lambda x: x.day)
df['start_hour'] = df.start_time.apply(lambda x: x.hour)

# end time
df.end_time = pd.to_datetime(df.end_time, format='%Y-%m-%d %H:%M:%S')
df['end_year'] = df.start_time.apply(lambda x: x.year)
df['end_month'] = df.end_time.apply(lambda x: x.month)
df['end_week'] = df.end_time.apply(lambda x: x.week)
df['end_day'] = df.end_time.apply(lambda x: x.day)
df['end_hour'] = df.end_time.apply(lambda x: x.hour)
```

```python
# Find the average number of users for each crossing above
bikes_avg = bikes_pt/bikes_pt_date_count

# Create a heatmap in seaborn
plt.figure(figsize=(8,8))
plt.title('Average number of users per hour and day of the week', fontsize = 12, fontweight = 'bold')
plt.tick_params(labelsize=12)
sns.heatmap(bikes_avg.round(), cmap='magma', linecolor='grey',linewidths=0.1, cbar=False, annot=True, fmt=".0f")
```

#### Observation:


- It seems rush hour starts from 4pm and peaks between 5-6pm
- Weekends (Saturday and Sunday) have fewer morning users compared to other days in the morning
- Weekends (Saturday and Sunday) have users well spread throughout the afternoon until peak hours
- Peak hours (5-6pm) are more dominated by users on weekdays

#### Number of Users in a Week

```python
# Checking the number of users throughout the week
plt.figure(figsize=(10,4))

df.groupby('weekday').size().plot(label = 'Users in a Week')
plt.title('Number of Users in a Week', fontsize = 12, fontweight = 'bold')
plt.legend(prop = {'size':10})
plt.tick_params(labelsize = 10)
```

#### Observation: 
- I feel strange about this result. In the previous heatmap, it can be seen that Saturday contributes a lot of users compared to Monday and Sunday since users are well spread throughout the day from noon to peak hour, but the line above shows a decrease on Saturday
- After figuring this out, I found one possible reason why this pattern occurs. As we know, the distinct values for Sunday and Monday are 5, which means there were 5 Mondays and Sundays that occurred in September 2013, but only 4 distinct values for Saturday. After checking the 2013 calendar, it turns out that only Mondays and Sundays occurred 5 times in September 2013. That's why the value count for Mondays and Sundays appears to be higher than any other days because they appeared 5 times in September 2013, not 4 times like the other days. However, apart from that, the heatmap gives more accurate results because the heatmap itself was created by dividing it based on its distinct values
- Further analysis: create a new df2 excluding 29 and 30 September 2013. Therefore, I tried to ensure the distinct values would be the same as the others, and therefore it would help reduce bias in my analysis (collective bias)

```python
# Create another dataframe that excludes 29 and 30 September 2013 --> using ~ to exclude the data
df2 = df.loc[~df['start_day'].isin([29, 30])]
df2
```
